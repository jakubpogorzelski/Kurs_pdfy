W pliku **dane.csv** znajdziesz dane kilku uczniów: imię, nazwisko, płeć, rok urodzenia, lista ocen.

Utwórz **jeden** plik `dane.pdf`, w którym każdy z uczniów będzie przedstawiony na osobnej stronie:

```
[Uczennica/Uczeń - wybierz jedno, na podstawie kolumny "płeć"] Imię i nazwisko - nagłówek
Rok urodzenia - normalny tekst

Oceny: - mniejszy nagłówek
Język Polski - 6
Język Angielski - 5
...
```

````python
from reportlab.lib.pagesizes import A2
from reportlab.platypus import SimpleDocTemplate, Paragraph, Spacer, Image
from reportlab.platypus import PageBreak
from reportlab.lib.styles import getSampleStyleSheet, ParagraphStyle
 
doc = SimpleDocTemplate("Oceny.pdf",pagesize=A4,rightMargin=72,
                        leftMargin=72,topMargin=72,bottomMargin=18)
styles = getSampleStyleSheet()
import pandas as pd

Story =[]
pageBreak = PageBreak()
df = pd.read_csv('oceny.csv',header=1, encoding = 'utf-8')
filtr = df.loc[1]
#Imię,Nazwisko,Płeć,Rok urodzenia,Język Polski,Język Angielski,Język Niemiecki,Matematyka,Chemia,Fizyka,Biologia,Historia,Geografia
for i in range(0,len(df.index)):
    imie = df.loc[i][0]
    nazwisko = df.loc[i][1]
    rok_ur = df.loc[i][3] 
    polski = df.loc[i][4]
    angol = df.loc[i][5]
    niemiec = df.loc[i][6]
    majza = df.loc[i][7]
    chemia = df.loc[i][8]
    fiza = df.loc[i][9]
    biola = df.loc[i][10]
    historia = df.loc[i][11]
    gegra = df.loc[i][12]
    if df.loc[i][2] == 'M':
        uczen = 'Uczeń'
    else: 
        uczen = 'Uczennica'
    
    naglowek = f'{uczen} - {imie} {nazwisko}'
    Story.append(Paragraph(naglowek,styles["Heading1"]))
    Story.append(Spacer(1, 12))

    ptext = f'{rok_ur}'
    Story.append(Paragraph(ptext))
    Story.append(Spacer(1, 12))

    ptext = f'Oceny:'
    Story.append(Paragraph(ptext,styles["Heading2"]))
    Story.append(Spacer(1, 12))
    
    ptext = f'Język Polski - {polski}'
    Story.append(Paragraph(ptext))
    Story.append(Spacer(1, 12))

    ptext = f'Język Angielski - {angol}'
    Story.append(Paragraph(ptext))
    Story.append(Spacer(1, 12))

    ptext = f'Język niemiecki - {niemiec}'
    Story.append(Paragraph(ptext))
    Story.append(Spacer(1, 12))

    ptext = f'Matematyka - {majza}'
    Story.append(Paragraph(ptext))
    Story.append(Spacer(1, 12))

    ptext = f'Chemia - {chemia}'
    Story.append(Paragraph(ptext))
    Story.append(Spacer(1, 12))

    ptext = f'Fizyka - {fiza}'
    Story.append(Paragraph(ptext))
    Story.append(Spacer(1, 12))

    ptext = f'Biologia - {biola}'
    Story.append(Paragraph(ptext))
    Story.append(Spacer(1, 12))

    ptext = f'Historia - {historia}'
    Story.append(Paragraph(ptext))
    Story.append(Spacer(1, 12))

    ptext = f'Geografia - {gegra}'
    Story.append(Paragraph(ptext))
    Story.append(Spacer(1, 12))

    Story.append(pageBreak)
    
    
    

doc.build(Story)
````
